```python
%matplotlib notebook 

import os

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
base_dir = '..'
data_path = os.path.join(base_dir,'data','raw','alabama_pm_aqs.csv')
```

```python
data = pd.read_csv(data_path)
```

```python
data.columns
```

```python
airnow_data = data[data['Source']=='AirNow']
aqs_data = data[data['Source']=='AQS']
```

```python
f,ax = plt.subplots(1,2,figsize=(5,2))
ax[0].hist(airnow_data['Daily Mean PM2.5 Concentration'].to_numpy())
ax[1].hist(aqs_data['Daily Mean PM2.5 Concentration'].to_numpy(),bins=100)
ax[1].set_xlim([0,30])
ax[0].set_xlim([0,30])
f.tight_layout()
```
